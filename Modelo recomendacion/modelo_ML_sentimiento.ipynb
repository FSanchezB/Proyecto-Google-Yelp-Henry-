```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

reviews_yelp = pd.read_csv(r'C:\Users\Fernando\Desktop\Proyecto-Google-Yelp-Henry-\Datasets\Datasets filtrados por categoria\reviews_final1.csv')
reviews_yelp2=pd.read_csv(r'C:\Users\Fernando\Desktop\Proyecto-Google-Yelp-Henry-\Datasets\Datasets filtrados por categoria\reviews_final2.csv')
reviews_yelp3=pd.read_csv(r'C:\Users\Fernando\Desktop\Proyecto-Google-Yelp-Henry-\Datasets\Datasets filtrados por categoria\reviews_final3.csv')

negocios_yelp = pd.read_csv(r'C:\Users\Fernando\Desktop\Proyecto-Google-Yelp-Henry-\Datasets\Tablas csv\negocios_yelp.csv' , encoding='utf-8')
ciudad = pd.read_csv(r'C:\Users\Fernando\Desktop\Proyecto-Google-Yelp-Henry-\Datasets\Tablas csv\Ciudad.csv', encoding='utf-8')
categoria = pd.read_csv(r'C:\Users\Fernando\Desktop\Proyecto-Google-Yelp-Henry-\Datasets\Tablas csv\Categoria.csv', encoding='utf-8')

reviews_yelp = reviews_yelp[reviews_yelp['text'].notna()]
reviews_yelp.columns = reviews_yelp.columns.str.strip()

reviews_yelp2 = reviews_yelp2[reviews_yelp2['text'].notna()]
reviews_yelp2.columns = reviews_yelp2.columns.str.strip()

reviews_yelp3 = reviews_yelp3[reviews_yelp3['text'].notna()]
reviews_yelp3.columns = reviews_yelp3.columns.str.strip()

negocios_yelp.columns = negocios_yelp.columns.str.strip()
```

```python
analyzer = SentimentIntensityAnalyzer()

# Función para analizar sentimiento con VADER
def analizar_sentimiento_vader(texto):
    score = analyzer.polarity_scores(texto)
    return score

# Aplicar VADER al análisis de sentimiento de las reseñas
reviews_yelp_sentimiento1= reviews_yelp['text'].apply(analizar_sentimiento_vader)
reviews_yelp_sentimiento1 = pd.DataFrame((reviews_yelp_sentimiento1).tolist())
reviews_yelp=pd.concat([reviews_yelp,reviews_yelp_sentimiento1],axis=1)
reviews_yelp
```

```python
analyzer = SentimentIntensityAnalyzer()

# Función para analizar sentimiento con VADER
def analizar_sentimiento_vader(texto):
    score = analyzer.polarity_scores(texto)
    return score

# Aplicar VADER al análisis de sentimiento de las reseñas
reviews_yelp_sentimiento2= reviews_yelp2['text'].apply(analizar_sentimiento_vader)
reviews_yelp_sentimiento2 = pd.DataFrame((reviews_yelp_sentimiento2).tolist())
reviews_yelp2=pd.concat([reviews_yelp2,reviews_yelp_sentimiento2],axis=1)
reviews_yelp2
```

```python
analyzer = SentimentIntensityAnalyzer()

# Función para analizar sentimiento con VADER
def analizar_sentimiento_vader(texto):
    score = analyzer.polarity_scores(texto)
    return score

# Aplicar VADER al análisis de sentimiento de las reseñas
reviews_yelp_sentimiento3= reviews_yelp3['text'].apply(analizar_sentimiento_vader)
reviews_yelp_sentimiento3 = pd.DataFrame((reviews_yelp_sentimiento3).tolist())
reviews_yelp3=pd.concat([reviews_yelp3,reviews_yelp_sentimiento3],axis=1)
reviews_yelp3
```

```python
reviews=pd.concat([reviews_yelp,reviews_yelp2,reviews_yelp3],axis=0)
reviews=reviews.dropna()
reviews
```

```python
reviews.drop(['neg','pos','neu'],axis=1,inplace=True)
#reviews.to_csv('reviews_sentimiento.csv',index=False,encoding='UTF8')
```

```python
# Solicitar entrada de ciudad y categoría
ciudad_input = input("Ingrese la ciudad: ")
categoria_input = input("Ingrese una de las siguientes categorias:\nFood/Restaurants\nHealth\nCar Services\nShopping\nHotel: ")

# Filtrar el ID de la ciudad basada en la entrada del usuario
ciudad_seleccionada = ciudad[ciudad['Ciudad'].str.lower() == ciudad_input.lower()]['Id_Ciudad'].values

if len(ciudad_seleccionada) == 0:
    print("Ciudad no encontrada.")
else:
    id_ciudad_seleccionada = ciudad_seleccionada[0]
    
    # Filtrar el ID de la categoría basada en la entrada del usuario
    categoria_seleccionada = categoria[categoria['Categoria'].str.lower() == categoria_input.lower()]['Id_Categoria'].values

    if len(categoria_seleccionada) == 0:
        print("Categoría no encontrada.")
    else:
        # Filtrar los negocios por ciudad y categoría
        negocios_filtrados = negocios_yelp[(negocios_yelp['Id_Ciudad'] == id_ciudad_seleccionada) & 
                                           (negocios_yelp['Id_Categoria'].isin(categoria_seleccionada))]

        if negocios_filtrados.empty:
            print("No se encontraron negocios para la categoría y ciudad ingresadas.")
        else:
            # Filtrar las reseñas de esos negocios
            reseñas_filtradas = reviews[reviews['business_id'].isin(negocios_filtrados['Id_Negocio'])]
            
            # Calcular el sentimiento promedio por negocio
            sentimiento_promedio = reseñas_filtradas.groupby('business_id')['compound'].mean().reset_index()
            
            # Unir con la tabla de negocios para obtener los nombres
            recomendaciones = pd.merge(sentimiento_promedio, negocios_filtrados, left_on='business_id', right_on='Id_Negocio')
            
            # Ordenar por sentimiento y mostrar resultados
            recomendaciones = recomendaciones.sort_values(by='compound', ascending=False)

            
            # Mostrar las recomendaciones
            if not recomendaciones.empty:
                print("\nRecomendaciones:")
                for index, row in recomendaciones.head(10).iterrows():
                    print(f"{row['Nombre']} - Dirección: {row['Direccion']} - Sentimiento: {row['compound']:.2f}")
            else:
                print("No hay suficientes datos de reseñas para generar recomendaciones.")
```
